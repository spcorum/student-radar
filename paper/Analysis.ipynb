```python
import os
import json
import matplotlib.pyplot as plt

import os

def plot_loss_curves_from_jsons(
    json_paths,
    save_path='combined_loss_plot.png',
    title='Generator and Discriminator Loss Curves: Student vs Original',
    show_model_names=True
):
    """
    Plot generator and discriminator loss curves for multiple models.
    """
    models = []

    label_counters = {
        "Student": 0,
        "Original": 0
    }

    label_suffixes = ['A', 'B', 'C', 'D', 'E', 'F'] # etc

    for path in json_paths:
        with open(path, "r") as f:
            data = json.load(f)
            model_type = data["meta"].get("model_type", "unknown").capitalize()
            model_name = data["meta"].get("model_name", os.path.basename(path))
            epochs = data["meta"].get("valid_epoch_numbers", [])
            g_loss = data["meta"].get("generator_loss_values", [])
            d_loss = data["meta"].get("discriminator_loss_values", [])

            if show_model_names:
                label = f"{model_type}GAN ({model_name})"
            else:
                count = label_counters[model_type]
                suffix = label_suffixes[count] if count < len(label_suffixes) else f"#{count+1}"
                label = f"{model_type}GAN-{suffix}"
                label_counters[model_type] += 1

            models.append({
                "label": label,
                "epochs": epochs,
                "g_loss": g_loss,
                "d_loss": d_loss
            })

    # Plotting
    plt.figure(figsize=(12, 6))

    # Generator losses first
    for model in models:
        plt.plot(model["epochs"], model["g_loss"], label=f"{model['label']} - Gen Loss", linestyle='-')

    # Discriminator losses next
    for model in models:
        plt.plot(model["epochs"], model["d_loss"], label=f"{model['label']} - Disc Loss", linestyle='--')

    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(title)
    plt.legend()
    plt.grid(True)

    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    plt.savefig(save_path, dpi=600)
    print(f"[Saved] Loss comparison plot to: {save_path}")
    plt.show()
    plt.close()

def fmt(x, float_precision=6):
    if isinstance(x, float):
        if abs(x) < 1e-6:
            return f"{x:.{float_precision}e}"
        else:
            return f"{x:.{float_precision}f}"
    elif x is None:
        return "N/A"
    else:
        return str(x)

def compare_models_from_json(results_paths, float_precision=6):
    # Load result files and format labels
    models = []
    for path in results_paths:
        with open(path, 'r') as f:
            data = json.load(f)
            model_type = data.get("meta", {}).get("model_type", "unknown").capitalize()
            model_name = data.get("meta", {}).get("model_name", os.path.basename(path))
            label = f"{model_type}GAN ({model_name})"
            models.append((label, data))

    # Collect all metric keys that follow {train, val, test} format
    metric_keys = []
    example = models[0][1]
    for key, value in example.items():
        if key == "meta":
            continue
        if isinstance(value, dict) and all(k in value for k in ["train", "val", "test"]):
            metric_keys.append(key)

    # For each metric, print a table with rows as data splits and columns as models
    for metric in metric_keys:
        print(f"\n=== Metric: {metric} ===\n")
        col_width = 28
        header = f"{'Split':<{col_width}}" + "".join([f"{label:<{col_width}}" for label, _ in models])
        print(header)
        print("-" * len(header))

        for split in ["train", "val", "test"]:
            row = f"{split:<{col_width}}"
            for _, result in models:
                value = result.get(metric, {}).get(split, None)
                row += f"{fmt(value, float_precision):<{col_width}}"
            print(row)
```

```python
results_paths = [
  '/content/drive/MyDrive/cs231n/student-radar/paper/data/analysis/student/model-9idubkqi/results.json',
  '/content/drive/MyDrive/cs231n/student-radar/paper/data/analysis/student/model-h1k2x95j/results.json',
  '/content/drive/MyDrive/cs231n/student-radar/paper/data/analysis/original/model-rghsxmf6/results.json',
  '/content/drive/MyDrive/cs231n/student-radar/paper/data/analysis/original/model-ltuwf57w/results.json'
]

plot_loss_curves_from_jsons(
    results_paths,
    save_path='/content/drive/MyDrive/cs231n/plots/combined_loss_plot_names.png',
    title='Generator and Discriminator Loss Curves: Student vs Original',
    show_model_names=True
)

plot_loss_curves_from_jsons(
    results_paths,
    save_path='/content/drive/MyDrive/cs231n/plots/combined_loss_plot_A_B.png',
    title='Generator and Discriminator Loss Curves: Student vs Original',
    show_model_names=False
)
```
